```python
from gradio_client import Client
import time

# Model endpoints
llama_url = "https://324dc1e5741615169f.gradio.live/"
gemma_url = "https://bc098720f84e74b598.gradio.live/"

llama_client = Client(llama_url)
gemma_client = Client(gemma_url)

def llama_chat(system_msg: str, user_prompt: str, timeout: int = 2000):
    """
    Calls Llama with chat formatting.
    """
    text_in = (
        "<|start_of_turn|>system\n" + system_msg + "<|end_of_turn|>\n"
        "<|start_of_turn|>user\n" + user_prompt + "<|end_of_turn|>\n"
        "<|start_of_turn|>assistant\n"
    )
    job = llama_client.submit(text_in)
    start = time.time()
    while not job.done() and time.time() - start < timeout:
        time.sleep(0.5)
    return job.outputs() if job.done() else None

def gemma_chat(user_prompt: str, timeout: int = 2000):
    """
    Calls Gemma with its specific chat formatting.
    """
    formatted_prompt = (
        "<start_of_turn>user\n" + user_prompt + "<end_of_turn>\n"
        "<start_of_turn>assistant\n"
    )
    job = gemma_client.submit(formatted_prompt)
    start = time.time()
    while not job.done() and time.time() - start < timeout:
        time.sleep(0.5)
    return job.outputs() if job.done() else None
```

```python
import gradio as gr
import time

def compare_models(system_msg, user_prompt, progress=gr.Progress(track_tqdm=True)):
    # Step 1: Call Llama
    progress(0.0, desc="Llama working...")
    llama_json = llama_chat(system_msg, user_prompt)
    progress(0.33, desc="Gemma working...")
    gemma_json = gemma_chat(user_prompt)
    
    # Step 2: Format outputs with Gemini
    llama_table = ""
    gemma_table = ""
    if llama_json:
        progress(0.5, desc="Formatting Llama output with Gemini...")
        llama_table = format_as_html_table(str(llama_json))
    if gemma_json:
        progress(0.66, desc="Formatting Gemma output with Gemini...")
        gemma_table = format_as_html_table(str(gemma_json))
    
    progress(1.0, desc="Done!")
    return llama_table, gemma_table

with gr.Blocks() as demo:
    gr.Markdown("## Compare Llama and Gemma Table Outputs")
    with gr.Row():
        system_msg = gr.Textbox(label="System Message", value="You are a marketing strategist specialized in creating market segmentation analyses for startups.")
        user_prompt = gr.Textbox(label="User Prompt", placeholder="Enter your prompt here...")
    submit_btn = gr.Button("Compare")
    progress_bar = gr.Markdown("")  # Placeholder for progress bar (handled automatically)
    
    with gr.Row():
        llama_html = gr.HTML(label="Llama Table")
        gemma_html = gr.HTML(label="Gemma Table")
    
    submit_btn.click(
        compare_models,
        inputs=[system_msg, user_prompt],
        outputs=[llama_html, gemma_html],
        api_name="compare",
        queue=True  # Enable queue for progress bars
    )

demo.launch()
```

```python
import time
import gradio as gr
from gradio_client import Client

def process_input(market_segmentation):
    # Step 1: Call Llama
    yield "", "", "Calling Llama..."
    llama_out = llama_chat(
        system_msg="You are a market segmentation assistant.",
        user_prompt=market_segmentation
    )
    yield llama_out or "<No response>", "", "Llama complete. Calling Gemma..."

    # Step 2: Call Gemma
    gemma_raw = gemma_chat(market_segmentation)
    yield llama_out or "<No response>", "", "Gemma complete. Formatting table..."

    # Step 3: Format HTML table via Gemini
    table_html = format_as_html_table(gemma_raw or "[]")
    yield llama_out or "<No response>", table_html, "Done."
```

```python
# Build Gradio UI
demo = gr.Blocks(css="""
.chat-input textarea {border-radius: 12px !important; height: 100px !important;}
""")

with demo:
    gr.Markdown("## Market Segmentation Analyzer")
    with gr.Column():
        input_box = gr.Textbox(
            placeholder="Enter market segmentation prompt...",
            lines=3,
            rounded=True,
            elem_id="chat-input"
        )
        progress = gr.Textbox(label="Progress", interactive=False)
    with gr.Row():
        llama_output = gr.Textbox(label="Llama Output", interactive=False)
        gemma_output = gr.HTML(label="Gemma Table Output")

    input_box.submit(fn=process_input,
                     inputs=[input_box],
                     outputs=[llama_output, gemma_output, progress])

if __name__ == "__main__":
    demo.launch()
```
